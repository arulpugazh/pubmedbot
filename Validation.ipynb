```python
import pandas as pd
import spacy
```

```python
df = pd.read_csv('valid/FARM_distilbert-base-uncased-distilled-squad_1_1.csv', 
                 skiprows=1, 
                 names=['id', 'article_id', 'question', 'answer', 'context', 'panswer',
       'pcontext', 'time', 'answer_similarity', 'context_similarity'])
```

```python
nlp = spacy.load('en_use_md')

for index, row in df.iterrows():
    answer = nlp(str(row['answer']))
    panswer = nlp(str(row['panswer']))
    ans_sim = panswer.similarity(answer)
    df.loc[index, 'answer_similarity'] = ans_sim
    
    context = nlp(str(row['context']))
    pcontext =nlp(str(row['pcontext']))
    ctx_sim = pcontext.similarity(context)
    df.loc[index, 'context_similarity'] = ctx_sim

df.to_csv('valid/test.csv')
```
